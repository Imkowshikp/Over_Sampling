```python
import pandas as pd;
import numpy as np;
from matplotlib import pyplot as plt;
import seaborn as sns;
```

```python
rawData = pd.read_csv("finalOutput_orden.csv")

rawData
```

```python
rawData.drop(rawData.columns[[0, 15]], axis = 1, inplace = True);

clientDataAltered = rawData;

clientDataAltered
```

```python

columns = clientDataAltered.columns.tolist()

columns = [c for c in columns if c not in ["Label"]]

target = "Label"

state = np.random.RandomState(42)

x = clientDataAltered[columns]

y = clientDataAltered[target]

```

```python

from imblearn.over_sampling import SMOTE

overSample = SMOTE(random_state = 42)

x_res, y_res = overSample.fit_resample(x, y)
```

```python
x_res.shape, y_res.shape
```

```python
from collections import Counter

print("Original dataset shape {}".format(Counter(y)))

print("Resampled dataset shape {}".format(Counter(y_res)))
```

```python
from imblearn.over_sampling import RandomOverSampler

os = RandomOverSampler(random_state = 42, sampling_strategy='auto', shrinkage=None)


x_train_res, y_train_res = os.fit_resample(x, y)
```

```python
x_train_res.shape, y_train_res.shape
```

```python
from collections import Counter

print("Original dataset shape {}".format(Counter(y)))

print("Resampled dataset shape {}".format(Counter(y_train_res)))
```

```python

from sklearn.model_selection import train_test_split;

x_train, x_test, y_train, y_test = train_test_split(x_train_res, y_train_res, test_size = 1/3, random_state = 6, stratify = y_train_res);
```

```python
from imblearn.over_sampling import RandomOverSampler

os = RandomOverSampler(random_state = 42, sampling_strategy='auto', shrinkage=None)


x_test_res, y_test_res = os.fit_resample(x_test, y_test)
```

```python
from sklearn.preprocessing import StandardScaler;

standardization = StandardScaler();

x_train = standardization.fit_transform(x_train);

x_test = standardization.transform(x_test);
```

```python
from sklearn.svm import SVC;

from sklearn.model_selection import GridSearchCV;



param_grid ={"kernel":["Linear", "poly", "rbf"], "C":[1.0, 10, 20, 50, 80, 100], 
             
             "degree":[1, 2, 3], "gamma":["auto", "scale"]
             
    
            }

grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3, n_jobs=-1, scoring="f1_weighted");

grid.fit(x_train, y_train);

```

```python
y_pred = grid.predict(x_test);
```

```python
from sklearn.metrics import confusion_matrix;

logisticRegressionConfusionMatrix = confusion_matrix(y_test, y_pred);

print(logisticRegressionConfusionMatrix);
```

```python
from sklearn.metrics import classification_report;

logisticRegressionClassifierReport = classification_report(y_test, y_pred);

print(logisticRegressionClassifierReport);
```

```python
from sklearn.metrics import accuracy_score;

SVMClassifierAccuracy = accuracy_score(y_test, y_pred);

print(SVMClassifierAccuracy);
```

```python
from sklearn.metrics import balanced_accuracy_score;

SVMClassifierBalanced = balanced_accuracy_score(y_test, y_pred);

print(SVMClassifierBalanced);
```
